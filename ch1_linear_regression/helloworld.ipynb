```python
from torch.autograd import Variable
```

```python
x=torch.ones(3)
x=Variable(x, requires_grad=True)
```

```python
y=(x**2)
z=y*3
print(z)
# grad=torch.Tensor([0.1, 1])
# z.backward(grad)
out=z.sum()
out.backward()
```
